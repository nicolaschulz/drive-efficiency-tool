# Efficiency Comparison and Energy Consumption Calculator for Drive Systems 
## According to <a href="https://www.dinmedia.de/de/norm/din-en-61800-9-1/279316566" target="_blank">EN61800-9-1</a>
This tool calculates the energy consumption of a drive system per cycle and per year, based on a load profile (*n(t)*, *M(t)*) and power loss data *P<sub>loss</sub>(n,M)*, both provided by the user.  
When providing loss data of a second drive system, the energy consumptions and energy costs of both systems can be compared.

```python
# EN 61800-9-2 Energieverbrauchs-Tool mit optionalem Lastprofil und Effizienzvergleich

import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output
import io
from scipy.interpolate import LinearNDInterpolator, griddata
import matplotlib.pyplot as plt
from scipy.ndimage import distance_transform_edt

# ---------- Hilfsfunktionen ----------
def parse_excel(uploaded_file):
    return pd.read_excel(io.BytesIO(uploaded_file["content"]))

def convert_speed_rpm_to_rads(df):
    df["Speed"] = df["Speed"] * 2 * np.pi / 60
    return df

def normiere_speed_torque(df, max_speed, max_torque):
    df["Speed_norm"] = df["Speed"] / max_speed
    df["Torque_norm"] = df["Torque"] / max_torque
    return df

def interpolate_losses(df_profile, ec_losses):
    required_cols = {"Speed (norm)", "Torque (norm)", "Power loss"}
    if not required_cols.issubset(ec_losses.columns):
        raise ValueError(f"Loss data must contain the columns  {required_cols}.")

    points = ec_losses[["Speed (norm)", "Torque (norm)"]].values
    values = ec_losses["Power loss"].values
    interpolator = LinearNDInterpolator(points, values, fill_value=np.nan)

    query_points = df_profile[["Speed_norm", "Torque_norm"]].values
    losses = interpolator(query_points)
    if np.any(np.isnan(losses)):
        print("‚ö†Ô∏è Attention: Some points are outside the valid interpolation range. Maximum value is used.")
        losses = np.nan_to_num(losses, nan=np.max(values))

    return losses

def convert_timestep(timestep_value, unit):
    if unit == "s":
        return timestep_value
    elif unit == "min":
        return timestep_value * 60
    elif unit == "h":
        return timestep_value * 3600
    else:
        raise ValueError("Invalid time unit")

def calculate_energy(df, losses, timestep_sec, energy_price):
    df = df.copy()
    df["P_mech"] = df["Speed"] * df["Torque"]
    df["P_in"] = df["P_mech"] + losses

    E_in = df["P_in"].sum() * timestep_sec / 3600_000
    E_out = df["P_mech"].sum() * timestep_sec / 3600_000
    losses_kwh = E_in - E_out
    cost = E_in * energy_price

    return E_in, E_out, losses_kwh, cost

def calculate_eta_array(ec_losses, max_speed_rpm, max_torque_nm):
    speed_norm = ec_losses["Speed (norm)"]
    torque_norm = ec_losses["Torque (norm)"]
    power_loss = ec_losses["Power loss"]

    speed_max_rads = max_speed_rpm * 2 * np.pi / 60
    torque_max = max_torque_nm

    omega = speed_norm * speed_max_rads
    torque = torque_norm * torque_max
    P_out = omega * torque
    P_in = P_out + power_loss
    eta = np.where(P_in > 0, P_out / P_in, 0) * 100
    return eta

def plot_efficiency_map(ec_losses, max_speed_rpm, max_torque_nm, title, vmin=None, vmax=None):
    fig, ax = plt.subplots(figsize=(8, 6))

    speed_norm = ec_losses["Speed (norm)"]
    torque_norm = ec_losses["Torque (norm)"]
    power_loss = ec_losses["Power loss"]

    speed_max_rads = max_speed_rpm * 2 * np.pi / 60
    torque_max = max_torque_nm
    speeds_percent = speed_norm * 100
    torques_percent = torque_norm * 100

    omega = speed_norm * speed_max_rads
    torque = torque_norm * torque_max
    P_out = omega * torque
    P_in = P_out + power_loss
    eta = np.where(P_in > 0, P_out / P_in, 0) * 100

    grid_x, grid_y = np.meshgrid(np.linspace(0, 110, 200), np.linspace(0, 110, 200))
    points = np.vstack([speeds_percent, torques_percent]).T
    grid_eff = griddata(points, eta, (grid_x, grid_y), method='linear')

    mask = np.isnan(grid_eff)
    distance, indices = distance_transform_edt(mask, return_indices=True)
    grid_eff[mask] = grid_eff[indices[0][mask], indices[1][mask]]

    contour = ax.contourf(grid_x, grid_y, grid_eff, levels=100, cmap="viridis", vmin=vmin, vmax=vmax)
    cbar = plt.colorbar(contour, ax=ax)
    cbar.set_label("Efficiency [%]", fontsize=12)

    ax.scatter(speeds_percent, torques_percent, c='red', edgecolors='black', s=100)
    for i in range(len(P_out)):
        label = f"P_out: {P_out[i]:.0f} W\nP_loss: {power_loss[i]:.0f} W\nEff: {eta[i]:.1f}%"
        x_text = min(speeds_percent[i] + 2, 98)
        y_text = min(torques_percent[i] + 2, 98)
        ax.text(x_text, y_text, label, fontsize=8, color='black')

    ax.set_xlim(0, 110)
    ax.set_ylim(0, 110)
    ax.set_xlabel("Speed [%]")
    ax.set_ylabel("Torque [%]")
    ax.set_title(title)
    ax.grid(True)
    plt.tight_layout()
    plt.show()

# ---------- GUI Elemente ----------
up_profile = widgets.FileUpload(accept='.xlsx', multiple=False)
up_profile_box = widgets.VBox([
    widgets.Label("üîº (Optional) Upload load profile with speed and torque (.xlsx):"),
    up_profile
])

ec_loss_1 = widgets.FileUpload(accept='.xlsx', multiple=False)
ec_loss_2 = widgets.FileUpload(accept='.xlsx', multiple=False)

timestep = widgets.FloatText(value=1.0, layout=widgets.Layout(width='100px'))
timestep_label = widgets.Label("time step", layout=widgets.Layout(width='160px'))
timestep_row = widgets.HBox([timestep, timestep_label])

timestep_unit = widgets.Dropdown(
    options=[("seconds", "s"), ("minutes", "min"), ("hours", "h")],
    value="h",
    layout=widgets.Layout(width='100px')
)
timestep_unit_label = widgets.Label("unit", layout=widgets.Layout(width='160px'))
timestep_unit_row = widgets.HBox([timestep_unit, timestep_unit_label])

max_speed = widgets.FloatText(value=3000, layout=widgets.Layout(width='100px'))
max_speed_label = widgets.Label("max speed [rpm]", layout=widgets.Layout(width='160px'))
max_speed_row = widgets.HBox([max_speed, max_speed_label])

max_torque = widgets.FloatText(value=17.8, layout=widgets.Layout(width='100px'))
max_torque_label = widgets.Label("max torque [Nm]", layout=widgets.Layout(width='160px'))
max_torque_row = widgets.HBox([max_torque, max_torque_label])

price = widgets.FloatText(value=0.2, layout=widgets.Layout(width='100px'))
price_label = widgets.Label("price [CHF/kWh]", layout=widgets.Layout(width='160px'))
price_row = widgets.HBox([price, price_label])

run_button = widgets.Button(description='Calculate', button_style='success')
output = widgets.Output()

# ---------- Berechnungslogik ----------
def on_run_clicked(b):
    with output:
        clear_output()
        print("üîç Start analysis...")

        if not ec_loss_1.value:
            print("‚ùå Please upload at least one set of efficiency data.")
            return

        ec1 = parse_excel(ec_loss_1.value[0])
        eta1 = calculate_eta_array(ec1, max_speed.value, max_torque.value)

        if ec_loss_2.value:
            ec2 = parse_excel(ec_loss_2.value[0])
            eta2 = calculate_eta_array(ec2, max_speed.value, max_torque.value)
            eta_all = np.concatenate([eta1, eta2])
            vmin, vmax = np.min(eta_all), np.max(eta_all)
        else:
            ec2 = None
            vmin, vmax = np.min(eta1), np.max(eta1)

        plot_efficiency_map(ec1, max_speed.value, max_torque.value, "Efficiency map system 1", vmin, vmax)

        if ec2 is not None:
            plot_efficiency_map(ec2, max_speed.value, max_torque.value, "Efficiency map system 2", vmin, vmax)

        if up_profile.value:
            df_profile = parse_excel(up_profile.value[0])
            print("‚úÖ Load profile loaded")
            df_profile = convert_speed_rpm_to_rads(df_profile)
            max_speed_rads = max_speed.value * 2 * np.pi / 60
            df_profile = normiere_speed_torque(df_profile, max_speed_rads, max_torque.value)

            losses1 = interpolate_losses(df_profile, ec1)
            timestep_sec = convert_timestep(timestep.value, timestep_unit.value)
            E1_in, E1_out, E1_loss, C1 = calculate_energy(df_profile, losses1, timestep_sec, price.value)

            print("\nüìä Results for system 1:")
            print(f"  Energy consumption: {E1_in:.2f} kWh")
            print(f"  Losses:         {E1_loss:.2f} kWh")
            print(f"  Energy cost:      {C1:.2f} CHF")

            if ec2 is not None:
                losses2 = interpolate_losses(df_profile, ec2)
                E2_in, E2_out, E2_loss, C2 = calculate_energy(df_profile, losses2, timestep_sec, price.value)

                print("\nüìä Results for system 2:")
                print(f"  Energy consumption: {E2_in:.2f} kWh")
                print(f"  Losses:         {E2_loss:.2f} kWh")
                print(f"  Energy cost:      {C2:.2f} CHF")

                print("\nüîÅ difference system 2 vs. system 1:")
                print(f"  Energy savings: {E1_in - E2_in:.2f} kWh")
                print(f"  Cost difference:  {C1 - C2:.2f} CHF")

                print("\nüìà Annual extrapolation (1 year):")
                stunden_pro_jahr = 8760
                t_stunden = timestep_sec / 3600
                zyklen_pro_jahr = stunden_pro_jahr / (len(df_profile) * t_stunden)

                E1_in_yr = E1_in * zyklen_pro_jahr
                E2_in_yr = E2_in * zyklen_pro_jahr
                E1_loss_yr = E1_loss * zyklen_pro_jahr
                E2_loss_yr = E2_loss * zyklen_pro_jahr
                C1_yr = C1 * zyklen_pro_jahr
                C2_yr = C2 * zyklen_pro_jahr

                print("  ‚ö° Energy consumption:  system 1: {:.1f} kWh | system 2: {:.1f} kWh".format(E1_in_yr, E2_in_yr))
                print("  üî• Losses:   system 1: {:.1f} kWh | system 2: {:.1f} kWh".format(E1_loss_yr, E2_loss_yr))
                print("  üí∞ Energy cost:    system 1: {:.2f} CHF | system 2: {:.2f} CHF".format(C1_yr, C2_yr))

                print("\nüü© Savings per year:")
                print("  ‚ö° Energy: {:.1f} kWh".format(E1_in_yr - E2_in_yr))
                print("  üî• Losses: {:.1f} kWh".format(E1_loss_yr - E2_loss_yr))
                print("  üí∞ Cost: {:.2f} CHF".format(C1_yr - C2_yr))

run_button.on_click(on_run_clicked)

# ---------- GUI anzeigen ----------
app = widgets.VBox([
    widgets.HTML("<h3>EN 61800-9-2 Energy calculation & efficiency comparison</h3>"),
    widgets.HTML("<b>Excel formats:</b><br>- Load profile: Columns ‚ÄòSpeed‚Äô [rpm], ‚ÄòTorque‚Äô [Nm]<br>- Loss data: Columns ‚ÄòSpeed (norm)‚Äô, ‚ÄòTorque (norm)‚Äô, ‚ÄòPower loss‚Äô [W]"),
    up_profile_box,
    widgets.Label("üîº Efficiency system 1 (.xlsx):"),
    ec_loss_1,
    widgets.Label("üîº Optional: Efficiency system 2 (.xlsx):"),
    ec_loss_2,
    max_speed_row,
    max_torque_row,
    timestep_row,
    timestep_unit_row,
    price_row,
    run_button,
    output
])

display(app)
```
